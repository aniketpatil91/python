## Who had more than the league average HRs in 2012 (filter out all players with less 500 PAs)? (Dataframe)

```python
j2 = df_2012[['lgID','HR']].groupby(['lgID'], as_index=False).mean()
```

```python
j2
```

```python
## joining 2 dataframes
merged= pd.merge(j1 , j2 ,on='lgID', how='inner')
```

```python
merged.head(5)
```

```python
## adding values to new columns that have average hr greater than league average
def que(x):
    if x['HR_x'] > x['HR_y']:
        return x['HR_x']
    else:
        ''
merged['que'] = merged.apply(que, axis=1)
```

```python
## removing null values
merged_removeNull =merged[~merged['que'].isnull()]
```

```python
df_players_avg = merged_removeNull[['playerID','nameFirst','nameLast']]
```

## Graph total HRs per season using bar graph

```python
df_grpby = batting[['yearID','HR']].groupby(['yearID'], as_index=False).sum()
x = df_grpby['yearID']
y = df_grpby['HR']
```

```python
fig = plt.figure()
fig.suptitle('Total HR per Season', fontsize=30)
fontsize2use = 15
font_label=20
font_title=25
plt.xticks(fontsize=fontsize2use)  
plt.yticks(fontsize=fontsize2use)
plt.rc('axes', labelsize=font_label)
plt.rcParams['figure.figsize']=(20,15)
plt.bar(x, y, align='center')
plt.xlabel('Year')
plt.ylabel('HR')
```
