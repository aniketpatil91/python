# Relative Strength Index

```python
def RSI(series, period):
 delta = series.diff().dropna()
 u = delta * 0
 d = u.copy()
 u[delta > 0] = delta[delta > 0]
 d[delta < 0] = -delta[delta < 0]
 u[u.index[period-1]] = np.mean( u[:period] ) #first value is sum of avg gains
 u = u.drop(u.index[:(period-1)])
 d[d.index[period-1]] = np.mean( d[:period] ) #first value is sum of avg losses
 d = d.drop(d.index[:(period-1)])
 rs = pd.stats.moments.ewma(u, com=period-1, adjust=False) / \
 pd.stats.moments.ewma(d, com=period-1, adjust=False)
 return 100 - 100 / (1 + rs)
 
```

```python
amzn_stk['RSI'] = RSI(amzn_stk['amzn'], 14)
googl_stk['RSI'] = RSI(googl_stk['googl'], 14)
ibm_stk['RSI'] = RSI(ibm_stk['ibm'], 14)
```

```python
plt.figure(figsize=(20,10))
plt.plot(amzn_stk['RSI'],marker='o')
plt.xlim('2017','2018')
plt.title('Relative Strenght Index (RSI) for Amazon(2017-2018)',fontsize=20)
plt.axhline(y=30,color = 'r',)
plt.axhline(y=70,color = 'r')
plt.legend(fontsize ='xx-large')
plt.xlabel('Years')
plt.ylabel('RSI')

plt.figure(figsize=(20,10))
plt.plot(googl_stk['RSI'],marker='o')
plt.xlim('2017','2018')
plt.title('Relative Strenght Index (RSI) for google(2017-2018)',fontsize=20)
plt.axhline(y=30,color = 'r',)
plt.axhline(y=70,color = 'r')
plt.legend(fontsize ='xx-large')
plt.xlabel('Years')
plt.ylabel('RSI')

plt.figure(figsize=(20,10))
plt.plot(ibm_stk['RSI'],marker='o')
plt.xlim('2017','2018')
plt.title('Relative Strenght Index (RSI) for IBM(2017-2018)',fontsize=20)
plt.axhline(y=30,color = 'r',)
plt.axhline(y=70,color = 'r')
plt.legend(fontsize ='xx-large')
plt.xlabel('Years')
plt.ylabel('RSI')
```

#### The Relative Strength Index (RSI) is a momentum indicator that compares the magnitude of recent gains and losses over a specified time period to measure speed and change of price movements of a security. It is primarily used to identify overbought or oversold conditions in the trading of an asset.

# Ease of Movement (EVM)

```python
amzn_stk = pd.read_csv('AMZN_2006-01-01_to_2018-01-01.csv')
amzn_stk['Date']=pd.to_datetime(amzn_stk['Date'])

googl_stk = pd.read_csv('GOOGL_2006-01-01_to_2018-01-01.csv')
googl_stk['Date']=pd.to_datetime(googl_stk['Date'])

ibm_stk = pd.read_csv('IBM_2006-01-01_to_2018-01-01.csv')
ibm_stk['Date']=pd.to_datetime(ibm_stk['Date'])

csco_stk = pd.read_csv('CSCO_2006-01-01_to_2018-01-01.csv')
csco_stk['Date']=pd.to_datetime(csco_stk['Date'])

aapl_stk = pd.read_csv('AAPL_2006-01-01_to_2018-01-01.csv')
aapl_stk['Date']=pd.to_datetime(aapl_stk['Date'])
aapl_stk = aapl_stk.set_index('Date')
msft_stk = pd.read_csv('MSFT_2006-01-01_to_2018-01-01.csv')
msft_stk['Date']=pd.to_datetime(msft_stk['Date'])
```

```python
def EVM(data, ndays): 
 dm = ((data['High'] + data['Low'])/2) - ((data['High'].shift(1) + data['Low'].shift(1))/2)
 br = (data['Volume'] / 100000000) / ((data['High'] - data['Low']))
 EVM = dm / br 
 EVM_MA = pd.Series(pd.rolling_mean(EVM, ndays), name = 'EVM') 
 data = data.join(EVM_MA) 
 return data 
```

```python
n = 14
AAPL_EVM = EVM(aapl_stk, n)
EVM = AAPL_EVM['EVM']

# Plotting the Price Series chart and the Ease Of Movement below
fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(2, 1, 1)
plt.plot(aapl_stk['Close'],lw=1)
plt.title('AAPLE Price Chart', fontsize=30)
plt.ylabel('Close Price')
plt.xlim('2017','2018')

plt.grid(True)
bx = fig.add_subplot(2, 1, 2)
plt.plot(EVM,'k',lw=0.75,linestyle='-',label='EVM(50)')
plt.legend(loc=2,prop={'size':9}, fontsize ='xx-large')
plt.ylabel('EVM values')
plt.title('14 Days EVM', fontsize=30)

plt.grid(True)
plt.xlim('2017','2018')
```

#### Ease of Movement (EMV) can be used to confirm a bullish or a bearish trend. A sustained positive Ease of Movement together with a rising market confirms a bullish trend, while a negative Ease of Movement values with falling prices confirms a bearish trend. The above chart shows bullish trend.
